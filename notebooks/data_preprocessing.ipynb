```python
def dates_below_date(col_series, date="2000-01-01"):
    col = col_series.copy()
    col = pd.to_datetime(col, format='%y%m', errors='coerce')
    mask = col >= pd.to_datetime(date)
    col[mask] -= pd.Timedelta(days=365.25*100)
    return col

def apply_plus_years(col_series, plus_years=20):
    col = col_series.copy()
    col += pd.Timedelta(days=365.25*plus_years)
    return col

def dbd_apy(col_series, date="2000-01-01", plus_years=20):
    col = dates_below_date(col_series, date)
    col = apply_plus_years(col, plus_years)
    return col
```
